```python
""" Plot and calculate DF/F0 etc for 2p data from a local injection of Ribo-jGCaMP8s
    processed in suite2p. """

# Location of suite2p output data.
datadir = "data/loc_ribo8s/"
name = "loc_ribo8s"
figname = datadir + name

# Load suite2p .npy output.
F = np.load(datadir + "F.npy", allow_pickle=True)
Fneu = np.load(datadir + "Fneu.npy", allow_pickle=True)
iscell = np.load(datadir + "iscell.npy", allow_pickle=True)

# Range of cells to plot for calciumplot over time plotmin, plotmax (seconds * acquisitionrate)
cellrange = [4, 15, 16, 24, 32]
plotmin, plotmax = int(40 * 30.9), int(140 * 30.9)

# Plot cells in cellrange over time plotmin to plotmax.
# calciumplot(plotmin=plotmin, plotmax=plotmax, cellrange=cellrange, figname=figname, F=F, Fneu=Fneu, iscell=iscell)

# Calculate DF/F0 etc for up to 80 cells if cellrange=False.
somaoverneuropil(cellrange=False, F=F, Fneu=Fneu, iscell=iscell, figname=figname)

# Live plot of calcium traces for browsing data.
# calciumplotlive(cellrange=[4, 15, 16, 24, 32], F=F, Fneu=Fneu, iscell=iscell)
```
